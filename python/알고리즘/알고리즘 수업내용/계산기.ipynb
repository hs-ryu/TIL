### 1-5.계산기 코드 -  isp, icp 미사용
- 약간만 코드의 양을 줄여 볼까요?

```
# isp, icp 사용하지 않음

def top(a):
    if len(a) == 0:
        return -1
    return a[len(a)-1]
 
 
def get_postfix_notation(math, T):
    # 2개의 빈 리스트를 사용하여 후위 표기법을 생성한다.
    stack = [] # 후위 표기법 문자를 담을 리스트
    operator = [] # 연산자만 담을 리스트
    idx = 0
    while idx < T:
        # +, - 연산자일 경우
        # operator에 있는 모든 연산자를 stack에 push한 후 자기 자신을 operator에 push한다
        if math[idx] == '+' or math[idx] == '-':
            if top(operator) == -1 or top(operator) == '(':
                operator += [math[idx]]
            else:
                stack.append(top(operator))
                operator.pop()
                continue
        # *, /  연산자일 경우
        # operator에 있는 *, / 연산자를 뒤에서부터 찾아 있을 경우, stack에 push하고
        # + 연산자를 만날 경우, 자기 자신을 operator에 push한다
        elif math[idx] == '*' or math[idx] == '/':
            if top(operator) == '*' or top(operator) == '/':
                stack.append(top(operator))
                operator.pop()
                continue
            else:
                operator += [math[idx]]
        elif math[idx] == ')':
            if top(operator) == '(':
                operator.pop()
            else:
                stack.append(top(operator))
                operator.pop()
                continue
        # 숫자나 ( 인 경우 stack에 push한다
        elif math[idx] == '(':
            operator.append(math[idx])
        else:
            stack.append(math[idx])
        idx += 1
 
    # operator에 남은 연산자를 뒤에서부터 push해준다.
    if len(operator):
        for i in range(len(operator)-1, -1, -1):
            stack.append(operator[i])
 
    return stack
 
 
# 후위 표기법 계산하는 함수
def get_sum(arr):
    cal = []
    for i in range(len(arr)):
        if arr[i] == '+':
            temp1 = top(cal)
            cal.pop()
            temp2 = top(cal)
            cal.pop()
            cal.append(temp2 + temp1)
        elif arr[i] == '-':
            temp1 = top(cal)
            cal.pop()
            temp2 = top(cal)
            cal.pop()
            cal.append(temp2 - temp1)
        elif arr[i] == '*':
            temp1 = top(cal)
            cal.pop()
            temp2 = top(cal)
            cal.pop()
            cal.append(temp2 * temp1)
        elif arr[i] == '/':
            temp1 = top(cal)
            cal.pop()
            temp2 = top(cal)
            cal.pop()
            cal.append(temp2 / temp1)
        else:
            cal.append(int(arr[i]))
 
    return cal[0]
 
 
for t in range(1, 11):
    N = int(input())
    math = input()
    postfix = get_postfix_notation(math, N)
    total = get_sum(postfix)
    print('#%d %s' % (t, total))
```

### 1-6. 계산기 코드 - rank 함수 구현

```
# calc 수정, rank를 dict로 수정
def rank(operator):
    if operator == '(':
        return 1
    elif operator == '+' or operator == '-':
        return 2
    elif operator == '*' or operator == '/':
        return 3
    elif operator == ')':
        return -1
    else:
        return 0
 
 
# 중위 표기 -> 후위 표기로 바꾸기
def postfix(calc_list):
    op = []  # 연산자 임시 저장
    result = []  # 후위연산으로 바꾼 결과 저장
    for i in calc_list:
        # 숫자면
        if not rank(i):
            # 그대로 스택에 저장
            result.append(i)
 
        # 연산자면
        else:
            # 여는 괄호면 무조건 op에 추가
            if rank(i) == 1:
                op.append(i)
 
            # 닫는 괄호면
            elif rank(i) == -1:
                for k in op:
                    # 여는 괄호가 나오면 여는괄호 pop해서 없애고 종료
                    if rank(op[-1]) == 1:
                        op.pop()
                        break
                    # 여는괄호 나올때까지 pop
                    else:
                        result.append(op.pop())
 
            # 괄호가 아닌 연산자면
            else:
                # 연산자 스택 비었으면 추가
                if not op:
                    op.append(i)
 
                # op에 들어있는 연산자와 우선 순위 비교
                else:
                    # 현재 연산자가 op에 들어있는 연산자보다 우선순위가 높으면 추가
                    if rank(i) > rank(op[-1]):
                        op.append(i)
 
                    # op에 있는 연산자가 현재 연산자와 같거나 크면
                    # op의 연산자를 빼서 result에 추가하고, op에 현재 연산자 추가
                    else:
                        result.append(op.pop())
                        op.append(i)
 
    # stack에 남아있는 연산자 추가
    for j in op:
        result.append(op.pop())
 
    return result
 
 
def calc(postfix):
    stack_calc = []
    # input 값 돌면서 확인
    for i in postfix:
        # 숫자면
        if not rank(i):
            # stack_calc에 추가
            stack_calc.append(int(i))
        # 연산자면
        else:
            # stack_calc에서 두 값을 꺼내서 연산
            b = stack_calc.pop()
            a = stack_calc.pop()
 
            if i == '*':
                total = a * b
            elif i == '+':
                total = a + b
            elif i == '-':
                total = a - b
            elif i == '/':
                total = a / b
            stack_calc.append(total)
 
    # 한번 돌고 stack에 남아있는 연산 실행
    while stack_calc:
        # 값이 하나면 종료
        if len(stack_calc) == 1:
            break
        b = stack_calc.pop()
        a = stack_calc.pop()
 
        if i == '*':
            total = a * b
        elif i == '+':
            total = a + b
        elif i == '-':
            total = a - b
        elif i == '/':
            total = a / b
        stack_calc.append(total)
 
    # 마지막 total 값이 계산 결과
    return total
 
for tc in range(1, 11):
    length = int(input())
    data = input()
 
    postfix_data = postfix(data)
    # print(postfix_data)
 
    print('#{} {}'.format(tc, calc(postfix_data)))
```
