## 1.최대값, 최소값 구하기

### 1-1. if 두 번 사용

```
# 배열 준비
import random
random.seed(0)
a = random.sample(range(1, 100), 10)
```

```
# min_value, max_value의 초기화를 범위 밖의 큰 값, 작은 값으로 하는 경우
# if 두 번 사용시 안전
# a가 1 ~100 범위의 수인 경우
min_value = 101  
max_value = 0
for i in range(len(a)):
    if min_value > a[i]:
        min_value = a[i]
    if max_value < a[i]:
        max_value = a[i]

print(min_value, max_value)
```

```
# if ~ elif를 사용했을때  a의 첫 번째 수가 최대값일 때 문제 발생
a = [100, 10, 10]
min_value = 101  
max_value = 0
for i in range(len(a)):
    if min_value > a[i]:
        min_value = a[i]
    elif max_value < a[i]:
        max_value = a[i]

print(min_value, max_value)
```

### 1-2. if ~ elif 사용

```
# min_value, max_value의 초기화를 a 배열의 첫 번째 값으로 하는 경우
# if ~ elif 사용할 수 있음
min_value = a[0]
max_value = a[0]
for i in range(1, len(a)):
    if min_value > a[i]:
        min_value = a[i]
    elif max_value < a[i]:
        max_value = a[i]

print(min_value, max_value)
```

```
# min_value, max_value의 초기화를 a 배열의 첫 번째 값으로 하는 경우
# a[0]가 최대값이어도 문제 없음
a = [100, 10, 10]
min_value = a[0]
max_value = a[0]
for i in range(1, len(a)):
    if min_value > a[i]:
        min_value = a[i]
    elif max_value < a[i]:
        max_value = a[i]

print(min_value, max_value)
```

### 1-3. get_maximum 함수 작성
- N개의 숫자를 argument로 전달하여 가장 큰 수를 return하는 함수 작성

```
def get_maximum(*args):
    print(args)
    max_num = 0
    for arg in args:
        if arg > max_num:
            max_num = arg
    return max_num
```

```
import random
random.seed(0)
a = random.sample(range(1, 100), 10)
print(get_maximum(*a))
```

```
x = get_maximum(4, 6, 2, 8, 3)
print(x)
```

## 5.슬라이딩 윈도우 기법
- 이해해 두면 좋은 문제 : https://bba-dda.tistory.com/5

```
import random
random.seed(0)

# N개의 숫자 중 연속된 M개의 합을 구했을 때 그 합이 가장 큰 것과 작은 것을 찾는 방법

N, M = 10, 3
a = random.sample(range(1, 100), N)
base = 0
for i in range(M):
    base += a[i]

minV = base
maxV = base
for i in range(0, N-M):
    base = base - a[i] + a[i+M]
    if minV > base:
        minV = base
    elif maxV < base:
        maxV = base

print(a, maxV, minV)
```

## 6.델타 사용으로 좌표구하기

### 6-1.상하좌우 좌표

```
dxy = [[-1, 0], [1, 0], [0, -1], [0, 1]]  # 상,하,좌,우
r, c = 1, 1
for dr, dc in dxy:
    nr = r + dr
    nc = c + dc
    # 이럴 때는 지나가면서 반복문 계속 돌리기 (continue)
    if nr < 0 or nr > 4 or nc < 0 or nc > 4:
        continue
    print(nr, nc)
```
