## 1.그래프 용어
- 그래프는 서로 관계가 있는 개체들과 그 관계를 그림으로 나타낸 것입니다. 이때 개체를 정점(Vertex) 그 관계를
간선(Edge)으로 나타냅니다.


image.png

- 왼쪽의 그래프는 정점이 7개이며, 간선이 8개인 그래프로 간선에 화살표 표시가 없으므로, 양방향으로 연결 된
상태입니다.
- 1-2와 2-1이 서로 연결된 구조이며, 어떻게 이동을 하더라도 동일한 비용이 요구되는 비가중치 그
래프 입니다. 
- 비가중치 그래프인 경우는 간선 위에 아무런 표시가 되어 있지 않습니다.
- 오른쪽의 그래프는 정점이 5개이며, 간선이 7개인 그래프로 간선에 화살표 표시가 있기 때문에 방향성을 갖고 연결 된 상태입니다. 
- 이를 방향 그래프라고 합니다. 
- 그리고 간선 위에 숫자가 써있는데 이것은 가중치(Weight)를 의미합니다. - 가중치란 정점에서 다른 연결된 정점으로 이동 시의 비용으로 해석할 수 있으며, 시간, 노력, 이익 등을 의미합니다.
- 정리해 보면 왼쪽의 그래프는 “비가중치 무방향 그래프”, 오른쪽은 “가중치 방향 그래프”가 되며, “비가중치 방향 그래프”와 “가중치 무방향 그래프”도 있을 수 있습니다.
- 참고로 모든 정점이 서로 서로 모두 연결되어 있는 그래프를 “완전 연결 그래프”라고 합니다.

## 2.그래프 데이터 저장

- 그래프의 저장에는 주로 2차원 배열 또는 Linked List를 사용하게 되는데, 알고리즘에서는 2차원 배열을 주로 사용합니다.

- 그래프를 위한 입력 데이터가 그림이 아닌 정점의 개수, 간선의 수, 간선의 목록으로 주어졌습니다.
- 이때는, 정점의 개수 만큼의 행, 열을 가진 2차원 배열을 준비하고, 행을 “출발지”, 열을 “도착지”로 하여 배열에 두 정점에 서로 관계가 있음을 표시합니다. 
- 반대로 행을 “도착지”, 열을 “출발지”로 하여 사용할 수도 있지만, 대

### 2-1. 비가중치 방향 그래프 
  > image.png

```
V, E = 5, 6
edges = [(1, 2), (2, 3), (1, 4), (2, 4), (1, 5), (3, 5)]
```

```
# make Adjacent Matrix
AM = [[0]*(V+1) for _ in range(V+1)]
for s, e in edges:
    AM[s][e] = 1
    
for x in AM:
    print(x)
```

```
# make Adjacent List
AL = [[] for _ in range(V+1)]

for s, e in edges:
    AL[s].append(e)
    
for x in AL:
    print(x)
```

## 3.재귀(reculsive)

### 3-1. 함수의 호출

```
# 함수의 호출은 stack의 원리 (LIFO)을 따른다
# 마지막에 호출된 함수가 먼저 종료된다.
```

```
def func_C():
    print('func_C1')
    print('func_C2')

def func_B():
    print('func_B1')
    func_C()
    print('func_B2')

def func_A():
    print('func_A1')
    func_B()
    print('func_A2')

func_A()
```

### 3-2. 재귀의 동작 이해
- 재귀 호출은 함수 내에서 자신을 다시 호출하는 것으로 실행 속도가 빠르다거나 메모리를 절약하는 방법은 아님
- 잘못 사용하면 RecursionError(Python)가 발생할 수 있음
- RecursionError : maximum recursion depth exceeded while calling a Python object
- 함수의 재귀 호출에는 depth 한계가 있음
   - 약 1000 번
   - sys.setrecursionlimit(숫자)를 사용해 한계를 변경할 수 있음
- 정확한 재귀 함수의 특징을 알고 적절
하게 사용해야 함
- 재귀 함수 내부의 코드 작성은 되도록 간단하고 명쾌하게 작성해야 함


```
# 코드의 실행
# 숫자가 한 줄에 하나씩 증가 되면서 출력되고 호출횟수의 한계를 넘어가면 에러가 발생함
```

```
import sys
sys.setrecursionlimit(1000)

def print_number(number):
    print(number, end = ' ')
    print_number(number+1)

print_number(1)
```

- 함수의 종료를 위한 코드가 필요함
- 종료를 위한 조건이 위에 작성되는 것이 일반적임

```
# 다음은 0 ~ 4까지 5개의 숫자를 재귀를 이용해 출력해 본 것이다.
# 다음 예제는 argument로 받은 숫자 level을 출력한 뒤, level + 1 하여 재귀 함수를 호출한다
# level >= N인 상황이면 함수를 종료한다
# 실행 화면을 보면 N이 5이기 때문에 0부터 4까지 출력되었다

N = 5
def func_A(level):
    if level >= N :
        return
    print(level, end=' ')
    func_A(level + 1)
    return

func_A(0)
```

## 4.부분집합 만들기

```
> 부분집합
   ijkl
0  0000
1  0001
2  0010
3  0011
4  0100
5
6
7
..
15 1111
```

### 4-1. 비트 연산 사용

```
# 비트 연산사용
N=4
subset = [0] * N
for i in range(1<<N):
    for j in range(N):
        subset[j] = 1 if i & (1<<j) else 0
    print(subset)      
```

### 4-2. 반복문 사용

```
# 반복문 사용 - for문 개수
subset=[0]*4
for i in range(2):
    subset[0] = i
    for j in range(2):
        subset[1] = j
        for k in range(2):
            subset[2] = k
            for l in range(2):
                subset[3] = l
                print(subset)  # 마지막에 할 동작
```

### 4-3. 재귀 사용

```
# 부분집합을 구할 목록의 item개수가 유동적인 경우 => 재귀로 풀자
```

```
1. item의 개수 => 재귀를 반복할 횟수
2. 재귀함수는 무엇을 할까?
```

### 5-1. N개 데이터에서 k개 데이터

#### 5-1-1. 반복문 사용

```
# N개 데이터에서 2개 데이터 뽑기
A = [1,3,5,7]

def func(N):
    for i in range(N-1):
        for j in range(i+1, N):
            print(i, j)  # 선택된 2개

func(len(A))        
```

```
# 위의 함수에서 N의 개수가 달라지는 경우 그대로 사용하지만,
# k의 개수가 달라지는 경우 for문을 추가/감소 해야한다
```

```
# N개 데이터에서 3개 데이터 뽑기
N = 5

def func(N):
    for i in range(N-2):
        for j in range(i+1, N-1):
            for k in range(j+1, N):
                print(i, j, k)  # 선택된 2개

func(N)   
```

#### 5-1-2. 재귀 사용
- 반복문의 사용은 k 변경시 for문의 사용에 변경이 있음
- 재귀를 사용하면, 재귀의 depth를 k로 제어하여 구현할 수 있음

```
# N개 데이터에서 2개 데이터 뽑기
N = 5
K = 3
A = [0]*K
def DFS_combi(L, S):
    if L == K:
        print(A)
        return
    
    for i in range(S, N-K+L+1):
        A[L] = i
        DFS_combi(L+1, i+1)
        A[L] = 0

DFS_combi(0, 0)        
```

### 5-2. 부분합 문제


```
# 1부터 12까지의 숫자를 원소로 가진 집합 A가 있다. 집합 A의 부분 집합 중 N개의 원소를 갖고 있고, 
# 원소의 합이 K인 부분집합의 개수를 출력하는 프로그램을 작성하시오.
# 해당하는 부분집합이 없는 경우 0을 출력한다. 모든 부분 집합을 만들어 답을 찾아도 된다.
# 예를 들어 N = 3, K = 6 경우, 부분집합은 { 1, 2, 3 } 경우 1가지가 존재한다.
```

```
# N개 데이터에서 K개 데이터 뽑기
DL = 12
N = 3
K = 6
A = [0]*N
# L : 몇 번째 뽑는 수
# st : 시작 숫자
# S : 뽑은 숫자의 합
def DFS_combi(L, st, S):
    if L == N:
        print(A)
        return
    
    for i in range(st, DL-N+L+2):
        if S + i > K : continue
        A[L] = i
        DFS_combi(L+1, i+1, S + i)
        A[L] = 0

# 0 번째 뽑는 수, 시작 숫자 1, 합 0
DFS_combi(0, 1, 0)        
```

### 6-3. 재귀 사용 (swap)

```
# 재귀 스왑 방식
arr = [1, 2, 3,4, 5]
N = len(arr)

def perm(idx):
    if idx == N:
        print(arr)
        return
    
    for i in range(idx, N):
        arr[idx], arr[i] = arr[i], arr[idx]
        perm(idx+1)
        arr[idx], arr[i] = arr[i], arr[idx]

perm(0)
```
