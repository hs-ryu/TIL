```
# 앞에서부터 접근
for t in range(1, int(input())+1):
    n = int(input())
    num_list = list(map(int, input().split()))
    length = len(num_list)
    from_idx = 0
    max_idx = 0
    money = 0
    while from_idx < n:
        max_cost = 0
        for l in range(from_idx, length):
            if num_list[l] >= max_cost:
                max_idx = l
                max_cost = num_list[l]
        for i in range(from_idx, max_idx):
            money += num_list[max_idx] - num_list[i]
        from_idx = max_idx + 1
    print("#%d %d" %(t, money))
```

## 4.재미있는 오셀로
- 델타의 사용
- changes의 사용으로 값을 가지고 있다가 원하는 상황에서 사용하고, 아니면 버릴 수 있도록 하는 기법

```
# 김영주님
# 상 하 좌 우 좌상 우상 좌하 우하
dx = [-1, 1, 0, 0, -1, -1, 1, 1]
dy = [0, 0, -1, 1, -1, 1, -1, 1]
 
def play_othello(board, m):
    for _ in range(m):
        y, x, color = map(int, input().split())
        board[x][y] = color
        for i in range(8):
            nx, ny = x+dx[i], y+dy[i]
            changes = []
            while board[nx][ny] == 3 - color:
                changes.append((nx, ny))
                nx += dx[i]
                ny += dy[i]
                if board[nx][ny] == 0:
                    break
            else:
                for a, b in changes:
                    board[a][b] = color
 
for t in range(1, int(input())+1):
    n, m = map(int, input().split())
    board = [[0]*(n+2) for _ in range(n+2)]
    half = n // 2
    board[half][half], board[half][half+1], board[half+1][half], board[half+1][half+1] = 2, 1, 1, 2
    play_othello(board, m)
    black, white = 0, 0
    for i in range(1, n+1):
        for j in range(1, n+1):
            if board[i][j] == 1:
                black += 1
            elif board[i][j] == 2:
                white += 1
    print('#%s %s %s' % (t, black, white))
```

## 6.농작물 수확하기

```
def get_solution(n, arr):
    m = n//2
    tot = 0
    # 커지는 부분까지
    for i in range(n//2 + 1):  
        s = m-i
        for j in range(s, s+(i*2+1)):
            tot += arr[i][j]

    for i in range(m+1, n):
        s = i-m
        for j in range(s, s+(n-i)*2-1):
            tot += arr[i][j]
    return tot


T = int(input())
for t in range(1, T+1):
    N = int(input())
    a = [list(map(int, input())) for _ in range(N)]
    sol = get_solution(N, a)
    print('#%d %d' % (t, sol))
```

## 7.스도쿠
- 배열을 사용해 중복없이 1 2 3 4 5 6 7 8 9 가 있음을 알아내는 것


### 7-1. set 사용

```
def get_result(puzzle):
    # 가로
    for row in puzzle:
        if len(row) != len(set(row)):
            return 0
    # 세로
    for col in zip(*puzzle):
        if len(col) != len(set(col)):
            return 0
    #격자
    for i in range(0, 7, 3):
        for j in range(0, 7, 3):
            square = puzzle[i][j:j + 3] + puzzle[i + 1][j:j + 3] + puzzle[i + 2][j:j + 3]
            if len(square) != len(set(square)):
                return 0
    return 1
 
T = int(input())
for t in range(1, T+1):
    arr = [list(map(int, input().split())) for _ in range(9)]
    ans = get_result(arr)
    print('#%d %d' % (t, ans))
```

### 7-2. list 사용

```
def get_solution(a):
    for i in range(0, 9, 3):
        for j in range(0, 9, 3):
            num = [0]*10
            for r in range(i, i+3):
                for c in range(j, j+3):
                    x = a[r][c]
                    num[x] += 1
                    if num[x] > 1:
                        return 0
    for r in range(9):
        num1 = [0] * 10
        num2 = [0] * 10
        for c in range(9):
            x = a[r][c]
            num1[x] += 1
            if num1[x] > 1:
                return 0
            x = a[c][r]
            num2[x] += 1
            if num2[x] > 1:
                return 0
    return 1


N = 9
T = int(input())
for t in range(1, T+1):
    a = [list(map(int, input().split())) for _ in range(N)]
    sol = get_solution(a)
    print('#%d %d' % (t, sol))
```

## 8.진기의 최고급 붕어빵

```
def my_bread(N, M, K, arrival):
    arrival.sort()
    for i in range(N):
        if (arrival[i] // M) * K < i+1:
            return 'Impossible'
    return 'Possible'
 
T = int(input())
for t in range(1, T + 1):
    N, M, K = map(int, input().split())
    arrival = list(map(int, input().split()))
    print('#%d %s' % (t, my_bread(N, M, K, arrival)))
```
